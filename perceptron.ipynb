```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load the dataset

```python
df = pd.read_csv('placement.csv')
```

```python
df
```

## Creating a scatter plot

```python
sns.scatterplot(x='cgpa', y='resume_score', hue='placed', data=df)
```

## Reassigning the colors

```python
palette = {
    0: 'red',
    1: 'blue' 
}
sns.scatterplot(x='cgpa', y='resume_score', hue='placed', palette=palette, data=df)
```
